- [A geodatabase is the most broad organizational group having a collection of geographic datasets of various types used to store and organize data and the data can be easily accessed. The feature dataset is a collection of featured classes that share the same coordinate system. The feature dataset supports data like topology and different terrain datasets. Featured classes are stored in the geodatabase and can be a stand alone or put into a featured dataset.They are a collection of common features like lines or polygons. They are like a folder with each having their own rolls or organizing data with geodatabase being the most broad then to feature dataset and the to feature classes where all the specific data is stored.]
